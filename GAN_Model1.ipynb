```python
import torch
from diffusers import StableDiffusionPipeline

# Load the model and move it to the available device (GPU/CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if device == "cuda" else torch.float32).to(device)

# Provide your prompt here
prompt = "boy looking out from the window"

# Generate the image based on the prompt
image = pipe(prompt).images[0]

display(image)  # Displays the image in the Colab notebook.
```
