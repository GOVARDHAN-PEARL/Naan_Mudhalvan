```python
import gradio as gr
from transformers import pipeline

# Load a pre-trained text classification model
classifier = pipeline("text-classification")

# Define the function to classify text
def classify_text(text):
    result = classifier(text)
    return result[0]['label'], result[0]['score']

# Set up Gradio interface
demo = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(placeholder="Enter text here..."),
    outputs=["text", "number"],
    title="Text Classification App",
    description="Enter some text to classify it into categories."
)

# Launch the app
demo.launch()
```

```python
import gradio as gr
from textblob import TextBlob
```

```python
def analyze_sentiment(text):
    analysis = TextBlob(text)
    sentiment = analysis.sentiment.polarity  # Returns a value between -1.0 (negative) and 1.0 (positive)
    if sentiment > 0:
        return "Positive"
    elif sentiment < 0:
        return "Negative"
    else:
        return "Neutral"

interface = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(lines=2, placeholder="Enter your text here..."),
    outputs="text",
    title="Sentiment Analysis",
    description="Enter a sentence to analyze its sentiment."
)

interface.launch(share=True)
```

```python
import gradio as gr
import spacy

# Load the SpaCy English model
nlp = spacy.load("en_core_web_sm")

def pos_tagging(text):
    # Process the input text with SpaCy
    doc = nlp(text)
    # Generate a list of tuples containing word and its POS tag
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags

# Create a Gradio interface
demo = gr.Interface(
    fn=pos_tagging,
    inputs="text",
    outputs="json",
    title="Part of Speech Tagging",
    description="Enter a sentence to get the part of speech tags for each word."
)

# Launch the interface
demo.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load the translation model (English to French in this example)
translator = pipeline("translation_en_to_fr")

def translate_text(text):
    # Translate the input text
    translation = translator(text, max_length=40)
    return translation[0]['translation_text']

# Create a Gradio interface
demo = gr.Interface(
    fn=translate_text,
    inputs="text",
    outputs="text",
    title="English to French Translator",
    description="Enter English text to get the French translation."
)

# Launch the interface
demo.launch()
```

```python
# Create a summarization pipeline
summarizer = pipeline("summarization", max_length=150, min_length=40, do_sample=False)

# Define a function that takes a text and returns a summary
def summarize(text):
    summary = summarizer(text)
    return summary[0]["summary_text"]

# Create a Gradio interface
interface = gr.Interface(
    fn=summarize,
    inputs=gr.Textbox(lines=10, label="Input Text"),
    outputs=gr.Textbox(label="Summary")
)

# Launch the interface
interface.launch()
```
